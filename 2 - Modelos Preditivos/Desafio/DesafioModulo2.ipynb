### Questão 9

```
# Aplique o modelo de normalização MinMaxScaler com os valores "default" para os dados de entrada. Qual é o menor valor para a variável "fixed acidity"
df_scaler = df.drop(columns='quality')
scaler = MinMaxScaler()
df_scaler = scaler.fit_transform(df_scaler)

# Retornando scaler a um modelo dataframe
colunas = list(df.columns)
colunas.remove('quality')
df_scaler = pd.DataFrame(data=df_scaler, columns=colunas)
df_scaler['fixed acidity'].min()
```

```
df_scaler.describe()
```

### Questão 10

```
# Aplicando o KNN
X = df_scaler
y = df.quality
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

clf_KNN = KNeighborsClassifier(n_neighbors=5)
clf_KNN.fit(X_train, y_train)
clf_KNN.score(X_test, y_test)
```

### Questão 11

```
#  Aplicando o algoritmo de árvore de decisão , qual é o valor da acurácia?
clf_arvore = DecisionTreeClassifier()
clf_arvore.fit(X_train, y_train)
clf_arvore.score(X_test, y_test)
```

### Questão 12

```
# Aplicando o algoritmo de floresta randomica
clf_floresta = RandomForestClassifier(max_depth=10, random_state=1)
clf_floresta.fit(X_train, y_train)
clf_floresta.score(X_test, y_test)
```

### Questão 13

```
# Aplicando o algoritmo SVM, qual é o valor da acurácia?
clf_svm = SVC(gamma='auto',kernel='rbf')
clf_svm.fit(X_train, y_train)
clf_svm.score(X_test, y_test)
```

### Questão 14

```
# Apicando a rede MLP, qual é o valor da acurácia?
clf_mlp = MLPClassifier( alpha=1e-5, hidden_layer_sizes=(5, 5), random_state=1)
clf_mlp.fit(X_train, y_train)
clf_mlp.score(X_test, y_test)
```

### Questão 15
